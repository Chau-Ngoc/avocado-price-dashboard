```python
import pandas as pd
import plotly.express as px
import plotly.io as pio
```

```python
pio.renderers.default = "notebook"
```

```python
df = pd.read_csv("avocado.csv")
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df["type"].unique()
```

```python
fig = px.line(
    data_frame=df[df["geography"] == "Albany"],
    x="date",
    y="average_price",
    color="type",
    title="Average Avocado Price over Years",
    labels={"date": "Years", "average_price": "Average Price", "type": "Type"},
)
fig.show()
```
